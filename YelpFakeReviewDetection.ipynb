```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
Yelp = pd.read_parquet("dataset/yelp_dataset.parquet")
Yelp
```

```python
Yelp_df = Yelp[['user_id','stars','useful_x','funny_x','cool_x', 'text','date','review_count','yelping_since','useful_y','funny_y','cool_y','friends','fans','average_stars']]
Yelp_df
```

```python
Yelp_df.rename(columns={'useful_x': 'useful_review','funny_x': 'funny_review','cool_x': 'cool_review','useful_y': 'useful_count','funny_y': 'funny_count','cool_y': 'cool_count'}, inplace=True)
Yelp_df
```

```python
Yelp_df['date'] = pd.to_datetime(Yelp_df['date'])
Yelp_df['Year_review'] = Yelp_df['date'].dt.year
Yelp_df
```
